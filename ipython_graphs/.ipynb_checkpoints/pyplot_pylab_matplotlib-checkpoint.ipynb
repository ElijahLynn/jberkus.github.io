```python
import pylab as plt
from matplotlib_venn import venn3, venn3_circles

mplset = set(['A', 'B', 'C', 'D', 'E'])
numpyset = set(['C','F','G'])
pypset = set(['A','B','C'])
plt.figure(figsize=(8,10))


v = venn3([mplset, numpyset, pypset], ('Matplotlib','Numpy','pyplot'))
for text in v.set_labels:
    text.set_fontsize(18)
plt.annotate('Pylab', xy = v.get_label_by_id('111').get_position(), xytext = (0,-70), size = 'xx-large',
            ha = 'center', textcoords = 'offset points', bbox = dict(boxstyle = 'round, pad = 0.5', fc = 'lime'),
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad = 0',color = 'gray'))

plt.show()
```
